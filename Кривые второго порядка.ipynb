```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

# 1 Эллипс
## В общем случае эллипс описывается уравнением $$\frac{x^2} {a^2}+\frac{y^2}{b^2}=1$$ где x и y - переменные, a и b - некоторые константы. Здесь мы будем иметь в виду a >= b, это значит, что эллипс будет "приплюснут" по оси Y. Чем больше отношение a/b, тем более сплюснутый будет эллипс.
## Важный случай эллипса возникает при a = b = R. Тогда эллипс превратится в окружность радиуса R, а уравнение эллипса - в уравнение окружности $$x^2+y^2=R^2$$

```python
a = [2, 3]
b = [1, 2]
theta = np.arange(0, 2*np.pi, 2*np.pi/100)
r0 = a[0]*b[0]/np.sqrt((a[0]*np.sin(theta))**2 + (b[0]*np.cos(theta))**2)
r1 = a[1]*b[1]/np.sqrt((a[1]*np.sin(theta))**2 + (b[1]*np.cos(theta))**2)
ax = plt.subplot(121, projection='polar')
ax.plot(theta, r0)
ax.set_rticks([0, 1, 2])
ax.set_title('a=2, b=1')
bx = plt.subplot(122, projection='polar')
bx.plot(theta, r1)
bx.set_rticks([0, 1, 2, 3])
bx.set_title('a=3, b=2')
plt.show()
```

# 2 Гипербола
## В общем случае гипербола описывается уравнением $$\frac{x^2}{a^2}-\frac{y^2}{b^2}=1$$ Точно также будем считать a >= b > 0. Пример гиперболы см. на рисунке
## В школьном курсе алгебры гиперболой называют кривую $$y=\frac{1}{x}$$ Это что, другая кривая? Нет, та же самая - если повернуть оси X и Y на 45 градусов, то гипербола при $$a = b = \frac{1}{\sqrt{2}}$$ переходит в $$y^{'}=\frac{1}{x^{'}}$$
#### Это легко показать, если выполнить преобразования поворота на 45 градусов: $$x=\frac{1}{\sqrt{2}}(x^{'}+y^{'}), y=\frac{1}{\sqrt{2}}(x^{'}-y^{'})$$Проделайте вычисления самостоятельно.

```python
theta = np.arange(0, 2*np.pi, 2*np.pi/100)
r = 1.0/np.sqrt(np.cos(theta)**2-np.sin(theta)**2)
ax=plt.subplot(111, projection='polar')
ax.plot(theta, r)
ax.set_rticks([0, 1, 2, 3])
ax.set_title('a=1, b=1')
plt.show()
```

# 3 Парабола
## С параболой в общем виде мы уже знакомы из школьного курса, в общем виде запишем: $$y=ax^2+bx+c$$В университетском курсе линейной алгебры параболу обычно обозначают по-другому: $$y^2=\beta{x}$$Это - та же парабола, повернутая на 90 градусов

```python
x = np.arange(-3., 3., 0.1)
ax=plt.subplot(121)
ax.plot(x, x**2-2)
ax.grid()
ax.set_title(r'$y=x^2-2$')
x = np.arange(0, 8, 0.1)
bx=plt.subplot(122)
bx.plot(x, np.sqrt(2*x), 'b', x, -np.sqrt(2*x), 'b')
bx.grid()
bx.set_title(r'$y^2=2x$')
plt.show()
```

# 4 Еще одна интересная фигура "второго" порядка получается из уравнения $$y^2=\beta^2x^2$$После извлечения квадратного корня получаем$$|y|=|\beta{x}|$$Это соответствует пересечению двух прямых

```python
x = np.arange(-4, 4, 0.1)
y = x/2
plt.plot(x, y, 'b', x, -y, 'b')
plt.grid()
plt.title(r'$y^2=\frac{x^2}{4}$')
plt.show()
```

# 6 Перейдем, наконец, к решению задач с параметром. Возьмем 18.1:
## Найти все значения a, при которых система уравнений имеет ровно 4 решения$$\begin{cases}x^4-y^4=12a-28 \\ x^2+y^2=a \end{cases}$$
## Решение: второе уравнение - это окружность с радиусом $$R^2=a, R=\sqrt{a}$$ Тогда имеем первое условие: a > 0, так как при a = 0 окружность вырождается в точку (0, 0), а при a < 0 - вообще исчезает.
## Посмотрим на первое уравнение - это кривая четвертого порядка, но простым преобразованием $$x^4-y^4=(x^2-y^2)(x^2+y^2)=a(x^2-y^2)$$мы упрощаем картину и получаем:$$\begin{cases}x^2-y^2=\frac{12a-28}{a} \\ x^2+y^2=a \\ a > 0 \end{cases}$$
## Теперь становится понятен смысл задачи - найти такие значения a, при которых окружность пересекается с гиперболой в 4 точках (гипербола - состоит из 2 половинок, см. выше). Причем, когда 12a-28 < 0 гипербола будет вертикальной, когда 12a-28=0 гипербола будет двумя прямыми, а когда 12a-28 > 0 гипербола будет горизонтальной. Иллюстрация ниже.

```python
theta=np.arange(0, 2*np.pi, 2*np.pi/100)
ax=plt.subplot(131, projection='polar')
a=1
r=[a]*len(theta)
g=np.sqrt((12*a-28)/(np.cos(theta)**2-np.sin(theta)**2))
ax.plot(theta, r, 'r', theta, g, 'g')
ax.set_title(u'a=1, нет пересечений')
bx=plt.subplot(132, projection='polar')
a=7.0/3
r=[np.sqrt(a)]*len(theta)
g=np.arange(0, 3, 0.1)
phi=[np.pi/4]*len(g)+[-np.pi/4]*len(g)+[3*np.pi/4]*len(g)+[-3*np.pi/4]*len(g)
g = np.array(4*[g]).flatten()
bx.plot(theta, r, 'r', phi, g, 'g')
bx.set_title(ur'$a=\frac{7}{3}$')
cx = plt.subplot(133, projection='polar')
a=3
r=[np.sqrt(a)]*len(theta)
g=np.sqrt((12*a-28)/(a*(np.cos(theta)**2-np.sin(theta)**2)))
cx.plot(theta, r, 'r', theta, g, 'g')
cx.set_title(ur'a=3, есть пересечения')
plt.show()
```

## 6 продолжение. Весь этот анализ графиков нужен для понимания смысла задачи и для того, чтобы не пропустить решение  a=7/3, при котором гипербола вырождается в прямые.
## Решаем задачу окончательно чисто алгебраически: сложим первое (приведенное) и второе уравнение, тогда $$x^2-y^2+x^2+y^2=\frac{12a-28}{a}+a \\ x^2=\frac{a^2+12a-28}{2a}$$Это уравнение разрешимо при $$a^2+12a-28>0 \\ (a-2)(a+14)>0 \\ (a < -14 \bigcup a>2) \bigcap a>0 \\ a>2$$Здесь знак строго больше взят потому, что нам нужно получить ДВА РАЗЛИЧНЫХ x (и два различных y, их комбинация и будет 4 искомых решения). Если разрешить a=2, тогда x=0 и 4 разных решения не получится. Аналогично, для анализа y, вычтем первое приведенное уравнение из второго:$$x^2+y^2-x^2+y^2=a-\frac{12a-28}{a} \\ y^2=\frac{a^2-12a+28}{2a} \\ a^2-12a+28>0 \\ (a-6+\sqrt{8})(a-6-\sqrt{8})>0 \\ 0 < a < 6-\sqrt{8} \bigcup a > 6 + \sqrt{8}$$Окончательно, с учетом допустимого диапазона для x, получаем: $$a \in (2, 6-\sqrt{8}) \bigcup (6+\sqrt{8}, +\infty)$$ Проверяем - вырожденное решение для a=7/3 входит в первый диапазон.
